```python
import numpy as np
import pandas as pd
```

## Train data

```python
train_df = pd.read_csv("../data/mnist_train.csv")
```

```python
train_df
```

```python
train_df.columns
```

```python
train_labels = (train_df["5"] != 0).astype("int")
```

```python
train_labels = train_labels.apply(lambda x: 1 if x == 1 else -1)
```

```python
train_labels = train_labels.to_numpy()
```

```python
train_labels = train_labels[:, None]
```

```python
train_labels.shape
```

```python
np.save("../data/train_labels.npy", train_labels)
```

```python
train_data = train_df.drop(["5"], axis=1).to_numpy() / 255.
```

```python
train_data = np.hstack([np.ones((train_data.shape[0], 1)), train_data])
```

```python
np.save("../data/train_data.npy", train_data)
```
